```python
#7. for the top 5 most popular artists. what are their most popular tracks? List the track name for each artist.
```

```python
import pandas as pd
df=pd.read_csv('C:\\Users\\krish\\OneDrive\\Desktop\\spotify.csv')
print(df.head())
average_popularity = df.groupby('Artist')['Popularity'].mean().reset_index()
top_5_artists = average_popularity.sort_values(by='Popularity', ascending=False).head(5)
artist_tracks = []
for artist in top_5_artists['Artist']:
    artist_data = df[df['Artist'] == artist]
    most_popular_track = artist_data.loc[artist_data['Popularity'].idxmax(), 'Track Name']
    artist_tracks.append((artist, most_popular_track))
print("Most Popular Tracks for the Top 5 Artists:")
for artist, track in artist_tracks:
    print(f"{artist}: {track}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\krish\\OneDrive\\Desktop\\spotify.csv')
print(df.columns)
if 'artist_name' not in df.columns or 'duration_ms' not in df.columns:
    print("Error: 'artist_name' or 'duration_ms' column not found in the data.")
else:
    sns.violinplot(
        x="artist_name",  
        y="duration_ms",  
        showmeans=True, 
        data=df
    )
    plt.xticks(rotation=45)
    plt.title("Distribution of Track Duration Across Artists")
    plt.xlabel("Artist Name")
    plt.ylabel("Track Duration (milliseconds)")
    plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv('C:\\Users\\krish\\OneDrive\\Desktop\\spotify.csv')
print(df.columns)
if 'artist_name' not in df.columns or 'popularity' not in df.columns:
    print("Error: 'artist_name' or 'popularity' column not found in the data.")
else:
    artist_data = df.groupby('artist_name')['popularity']
    sns.swarmplot(
        x=artist_data.index,  
        y=artist_data.values, 
        data=artist_data
    )
    plt.xticks(rotation=45)
    plt.title("Distribution of Track Popularity Across Artists")
    plt.xlabel("Artist Name")
    plt.ylabel("Track Popularity")
    plt.show()
```
